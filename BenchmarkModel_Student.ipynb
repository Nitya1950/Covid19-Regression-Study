```python
from sklearn.utils import shuffle 
 
# Shuffling the Dataset 
data = shuffle(data, random_state = 42) 
 
#creating 4 divisions 
div = int(data.shape[0]/4) 
 
# 3 parts to train set and 1 part to test set 
train = data.loc[:3*div+1,:] 
test = data.loc[3*div+1:]
```

```python
train.head()
```

```python
test.head()
```

```python

test['simple_mean']=train['percentage_mean'].mean()
```

```python
from sklearn.metrics import mean_absolute_error as mae
error= mae(test['percentage_mean'],test['simple_mean'])
```

```python
error
```

```python
avg_per = pd.pivot_table(train, values='percentage_mean', index = ['Avg_time_studying'], aggfunc=np.mean) 
avg_per
```

```python
# initializing new column to zero 
test['avg_per_mean'] = 0 
 
# For every unique entry in Outlet_Identifier 
for i in train['Productivity'].unique(): 
  # Assign the mean value corresponding to unique entry 
  test['avg_per_mean'][test['Productivity'] == str(i)] = train['percentage_mean'][train['Productivity'] == str(i)].mean()

```

```python
#calculating mean absolute error 
avg_per_error = mae(test['percentage_mean'] , test['avg_per_mean'] ) 
avg_per_error
```

```python
combo = pd.pivot_table(train, values = 'percentage_mean', index = ['Productivity','Avg_time_studying'], aggfunc = np.mean) 
combo
```

```python
# Initiating new empty column 
test['Super_mean'] = 0 
 
# Assigning variables to strings ( to shorten code length) 
s2 = 'Productivity' 
s1 = 'Avg_time_studying' 
 
# For every Unique Value in s1 
for i in test[s1].unique(): 
  # For every Unique Value in s2 
  for j in test[s2].unique(): 
    # Calculate and Assign mean to new column, corresponding to both unique values of s1 and s2 simultaneously 
    test['Super_mean'][(test[s1] == i) & (test[s2]==str(j))] = train['percentage_mean'][(train[s1] == i) & (train[s2]==str(j))].mean()
```

```python
#calculating mean absolute error 
super_mean_error = mae(test['percentage_mean'] , test['Super_mean'] ) 
super_mean_error
```
